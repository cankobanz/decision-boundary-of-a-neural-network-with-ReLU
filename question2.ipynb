# Training the model

```python
from sklearn.metrics import accuracy_score
```

```python
from  sklearn.neural_network import MLPClassifier #to use neural net
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
architecture = (4, 8, 1) ###(A num_layers sized tuple with number of hidden neurons as each element)
activationf = 'relu'
learning_rate=0.01
```

```python
mlp = MLPClassifier(hidden_layer_sizes=architecture,max_iter=1000,activation=activationf,learning_rate_init=learning_rate)
mlp.fit(X_train,y_train) #training classifier
```

```python
prediction =  mlp.predict(np.array(X_test))
df = pd.DataFrame(
    {
        "Prediction": prediction,
        "Real": y_test,
    }
)
df
```

```python
print(accuracy_score(y_test,prediction))
```

# Decision Boundary Plotting

```python
xx=np.linspace(-2.5, 2.5, 1000) #bounds of data
yy=np.linspace(-2.5, 2.5, 1000)

X11=[]
X22=[]
pred=[]
for point1 in xx:
    for point2 in yy:
        X11.append(point1)
        X22.append(point2)
        pred.append(mlp.predict(np.array([point1,point2]).reshape(1,-1))) #getting predictions across each point
```

```python
# Plotting decision boundary
f, ax1 = plt.subplots(figsize=(8, 6))
color= ['yellow' if l == 1 else 'purple' for l in pred]
ax1.scatter(X11, X22, c=color, s=50, linewidth=1)

# Plotting points
color2= ['blue' if l == 1 else 'green' for l in y_train]
ax1.scatter(X_train[:,0], X_train[:, 1], c=color2, s=100,
            cmap="RdBu", vmin=-.2, vmax=1.2,
            edgecolor="white", linewidth=1)
```
